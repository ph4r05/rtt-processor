```python
import os
import glob
import shutil
import json
import math
```

```python
dest_size = 200000000
files = glob.glob('/tmp/retest-200mb/*_100MiB_*.json')
for fname in files:
    js = json.load(open(fname))
    prevsize = js['tv_size'] * js['tv_count']
    if dest_size != dest_size:
        continue
    js['tv_count'] = int(math.ceil(dest_size / float(js['tv_size'])))
    newsize = js['tv_size'] * js['tv_count']
    nfname = fname.replace('_100MiB_', '_200MiB_')
    json.dump(js, open(nfname, 'w+'))
```

```python
files = glob.glob('/tmp/retest-200mb/*_100MiB_*.json')
for f in files:
    os.unlink(f)
```

```python
files = glob.glob('/tmp/retest-200mb/*_200MiB_*.json')
for f in files:
    nfile = f.replace('SECMARGINPAPER25_', 'SECMARGINPAPER79_')
    nfile = nfile.replace('SECMARGINPAPER26_', 'SECMARGINPAPER79_')
    nfile = nfile.replace('SECMARGINPAPER27_', 'SECMARGINPAPER79_')
    os.rename(f, nfile)
    js = json.load(open(nfile))
    js['file_name'] = os.path.splitext(os.path.basename(nfile))[0] + '.bin'
    json.dump(js, open(nfile, 'w+'))
```

```python
files = glob.glob('/tmp/retest-200mb/*_200MiB_*.json')
lines = []
genl = []
for f in files:
    bname = os.path.basename(f)
    expname = os.path.splitext(bname)[0]+'.bin'
    lines.append('submit_experiment --all_batteries -c "../sample_configs/100MB.json" -f {0} -n {0}'.format(expname))
    genl.append('crypto-streams-v3.0 -c=' + bname)
    
print('\n'.join(lines))
with open('/tmp/retest-200mb/submit.sh', 'w+') as fh:
    fh.write('\n'.join(lines))
with open('/tmp/retest-200mb/generate.sh', 'w+') as fh:
    fh.write('\n'.join(genl))
```
